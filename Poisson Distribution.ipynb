```python
import seaborn as sns
```

```python
poisson_data = random.poisson(lam =20,size = 200)
sns.histplot(poisson_data,label = 'poisson',kde = 'False',color ='green')
uniform_data = random.uniform(size = 200)
sns.histplot(uniform_data,label='uniform',kde = 'False',color = 'red')
binomial_data = random.binomial(n=20,p=0.5,size =200)
sns.histplot(binomial_data,label='binomial',kde='False',color = 'purple')
normal_data = random.normal(loc = 20 ,scale = 20 ,size = 200)
sns.histplot(normal_data,label='normal',kde = 'False',color = 'blue')
plt.legend()
plt.title('THE HISTOGRAM SHOWING POISSON,UNIFORM,BINOMIAL,AND NORMAL DISTRIBUTION GRAPHS',color ='PURPLE')
plt.xlabel('VALUE',color= 'purple')
plt.ylabel('DENSITY',color = 'purple')
```

```python
poisson_data = random.poisson(lam =20,size = 2000)
sns.kdeplot(poisson_data,label = 'poisson',color ='green')
uniform_data = random.uniform(size = 2000)
sns.kdeplot(uniform_data,label='uniform',color = 'red')
binomial_data = random.binomial(n=20,p=0.5,size =2000)
sns.kdeplot(binomial_data,label='binomial',color = 'purple')
normal_data = random.normal(loc = 20 ,scale = 20 ,size = 2000)
sns.kdeplot(normal_data,label='normal',color = 'blue')
plt.legend()
plt.title('THE HISTOGRAM SHOWING POISSON,UNIFORM,BINOMIAL,AND NORMAL DISTRIBUTION GRAPHS',color ='PURPLE')
plt.xlabel('VALUE',color= 'purple')
plt.ylabel('DENSITY',color = 'purple')
```
