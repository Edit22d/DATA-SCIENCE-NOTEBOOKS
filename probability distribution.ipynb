## NORMAL DISTRIBUTION

## Generate a random normal distribution of size 2*3

```python
from numpy import random
x=random.normal(size=(2,3))
print(x)
```

```python
from numpy import random
x=random.normal(size=(1,1))
print(x)
```

## Generate a random normal distribution of 2*3 with mean at 1 and standard deviation of 2

```python
from numpy import random
x=random.normal(loc = 1,scale = 2,size =(2,3))## loc is the mean,scale is the standard deviation,
print(x)
```

```python
from numpy import random
x=random.normal(loc =4,scale =6,size =(4,6))
print(x)
```

## Visualization of normal distribution

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.distplot(random.normal(size=1000),hist=False)
plt.show()             
```

```python
data = random.normal(size=100)
sns.kdeplot(data,bw_adjust=0.5)
plt.show()
```
